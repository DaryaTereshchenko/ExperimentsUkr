```python
import pandas as pd 
from statsmodels.stats import inter_rater as irr
import numpy as np 
```

```python
corpus = pd.read_csv(r"C:\Users\darias\Desktop\UKR\data\Kappa.csv", encoding="utf-8")
c = corpus.iloc[1: , :]
c
```

```python
corpus['sum_1'] = corpus[['Column2','Column3','Column4', 'Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11', 'Column12',
                  'Column13', 'Column14', 'Column15']].sum(axis=1)
corpus['sum_0'] = (corpus[['Column2','Column3','Column4', 'Column5', 'Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11', 'Column12',
                  'Column13', 'Column14', 'Column15']]==0).sum(axis=1)
```

```python
data = corpus.iloc[1: , :]
data = data.loc[:, ['sum_1', 'sum_0']].to_numpy()
data
```

```python
irr.fleiss_kappa(data, method='fleiss')
```
